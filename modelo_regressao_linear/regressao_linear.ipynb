```python
import pandas as pd
dados = pd.read_csv('celsius.csv')
```

```python
dados.info()
dados.head()
```

```python
import seaborn as sb

sb.scatterplot(x="celsius", y="fahrenheit", data=dados)
```

```python
sb.scatterplot(x="celsius", y="fahrenheit", data=dados, hue="fahrenheit", palette="coolwarm")
```

```python
#Caracteristicas (x), etiquetas (y)
X = dados['celsius']
Y = dados['fahrenheit']
```

```python
#X.values
#[-40, -10, 0, 8, ..] -> forma em que os dados estao apresentados
#[[-40], [-10], [0], [8], ..] -> forma em que devem ser apresentados para aplicar o modelo

x_processado = X.values.reshape(-1,1)
y_processado = Y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression 

modelo = LinearRegression()

modelo.fit(x_processado, y_processado)
```

```python
#realizando uma prediccao

celsius = [[23]]
prediccao = modelo.predict(celsius)

print(f'{celsius}º celsius = {prediccao}º fahrenheits')
```

```python
#Avaliando a precisao do modelo
#de 0 a 1 (excelente)
modelo.score(x_processado, y_processado) #este modelo eh perfeito pois a prediccao eh linear
```
